```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
data = load_iris()
X = data.data
y = data.target

X_fp32 = X.astype(np.float32)
X_fp64 = X.astype(np.float64)

diff = np.abs(X_fp64 - X_fp32)

print("Mean Precision Loss:", np.mean(diff))
```

```python
plt.plot(diff)
plt.title("Floating Point Precision Drift")
plt.xlabel("Feature Index")
plt.ylabel("Error")
plt.show()
```
